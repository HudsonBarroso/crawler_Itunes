```python
from lxml import html
import requests
import time

class AppCrawler:
    def __init__(self, url_inicial, nivel_max):
        self.url_inicial = url_inicial
        self.nivel_max = nivel_max
        self.nivel_atual = 0
        self.links_nivel = []
        self.apps = []

    def get_app_do_link(self, link):
        pagina_inicial = requests.get(link)
        arvore = html.fromstring(pagina_inicial.text)
        nome = arvore.xpath('//h1[@class="product-header__title app-header__title"]/text()')[0].strip()
        desenvolvedor = arvore.xpath('//dd[@class="information-list__item__definition"]/text()')[0].strip()
        tamanho = arvore.xpath('//dd[@class="information-list__item__definition"]/text()')[1].strip()
        links = arvore.xpath('//a[@class="we-lockup targeted-link l-column small-2 medium-3 large-2 we-lockup--shelf-align-top"]/@href')
        app = App(nome, desenvolvedor, tamanho, links)
        return app

    def crawl(self):
        app = self.get_app_do_link(self.url_inicial)
        self.apps.append(app)
        self.links_nivel.append(app.links)
        while self.nivel_atual < self.nivel_max:
            links_atuais = []
            for link in self.links_nivel[self.nivel_atual]:
                app_atual = self.get_app_do_link(link)
                links_atuais.extend(app_atual.links)
                self.apps.append(app_atual)
                time.sleep(2)
            self.nivel_atual += 1
            self.links_nivel.append(links_atuais)
        return

count = 0
class App: 
    def __init__(self, nome, desenvolvedor, tamanho, links):
        self.nome = nome
        self.desenvolvedor = desenvolvedor
        self.tamanho = tamanho         
        self.links = links
        

    def __str__(self):            
        return (str(count) + ".The game name: " + self.nome +
                "\n  Game size: " + self.tamanho +
                "\n  Developer name: " + self.desenvolvedor + "\n") 

crawler = AppCrawler('https://apps.apple.com/br/app/clash-of-clans/id529479190', 1)
crawler.crawl()

file1 = open("crawler_info.txt","w")
file1.write("Crawler to access the link: " + crawler.url_inicial + "\n")
for app in crawler.apps:
    count += 1    
    file1.writelines(str(count) + ".The game name: " + app.nome +
                     "\n  Game size: " + app.tamanho +
                     "\n  Developer name: " + app.desenvolvedor + "\n\n")
    print(app)

file1.close()  
    
```
